```python
def getMonthDelta(snapshot, mth):
    return int((mth - mth%100)/100 - (snapshot - snapshot%100)/100)*12 + (mth%100 - snapshot%100)
```

```python
df['months_until_checkoff'] = df.apply(lambda row: getMonthDelta(row['snapshot'], row['mth_code']), axis=1)
```

```python
macro = macro[:][7:]
macro
macro['year'] = macro['Mnemonic'].str[-4:]
macro['year'] = pd.to_numeric(macro['year'])
macro.head()
```

```python
df['year'] = df['snapshot'].astype(str).str[:4]
df.head()
```

```python
def convert_year_to_str(year, month):
    if int(month) < 10:
        return f'{year}0{month}'
    else:
        return f'{year}{month}'
```

```python
convert_year_to_str(2018,9)
```

```python
macro = macro[(macro['year'] == 2018) | (macro['year'] == 2019) | (macro['year'] == 2020) | (macro['year'] == 2021)]
month = macro['Mnemonic'].str.split("/", n=2, expand=True)
macro["month"] = month[0]
macro['code'] = macro.apply(lambda row: convert_year_to_str(row['year'], row['month']), axis=1)
macro
```

```python
gdp_map = {}
for row in macro['code']:
    gdp_map[row] = float(macro[macro['code'] == row]['M_FGDP.IUSA'])
print(gdp_map)
```

```python
income_map = {}
for row in macro['code']:
    income_map[row] = float(macro[macro['code'] == row]['M_FYLE.IUSA'])
print(income_map)
```

```python
debt_service_map = {}
for row in macro['code']:
    debt_service_map[row] = float(macro[macro['code'] == row]['M_FSDEBT.IUSA'])
print(debt_service_map)
```

```python
unemployment_map = {}
for row in macro['code']:
    unemployment_map[row] = float(macro[macro['code'] == row]['M_FUNI.IUSA'])
print(unemployment_map)
```

```python
df['Home Price'] = df['snapshot'].astype(str).map(home_price_map)
df['Debt Service'] = df['snapshot'].astype(str).map(debt_service_map)
df['Income Personal'] = df['snapshot'].astype(str).map(income_personal_map)
df['Interest Rate'] = df['snapshot'].astype(str).map(interest_rate_map)
df['Consumer Confidence Index'] = df['snapshot'].astype(str).map(confidence_consumer_map)
df['Dow Jones'] = df['snapshot'].astype(str).map(dow_jones_map)
df['Unemployment'] = df['snapshot'].astype(str).map(unemployment_map)
#
df['Interest Rate 10'] = df['snapshot'].astype(str).map(interest_rate_10_map)
df.head()
```

```python
feature_list = ['GDP', 'Personal Saving', 'Fixed Investment R', 'Income', 'Home Price', 'Debt Service', 'Income Personal', 'Consumer Confidence Index', 'Dow Jones', 'Unemployment', 'Interest Rate 10']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.naive_bayes import GaussianNB
from sklearn import tree
from sklearn.model_selection import train_test_split

X = df[feature_list]
```

```python
from sklearn import preprocessing
X_train, X_test, y_train, y_test = train_test_split(X, df['months_until_checkoff'], test_size=0.2)
scaler = preprocessing.StandardScaler().fit(X_train)
X_scaled = scaler.transform(X_train)
```

```python
df[feature_list].corrwith(df['months_until_checkoff'])
```

```python
clf =  tree.DecisionTreeRegressor().fit(X_scaled, y_train)
```

```python
X_test_scaled = scaler.transform(X_test)
print(f'score: {clf.score(X_test_scaled, y_test)}')
from sklearn.metrics import accuracy_score
#y_p = clf.predict(X_test_scaled)
#accuracy_score(y_p, y_test)
pd.Series(clf.predict(X_test_scaled)).unique()
```
