```python
high_corr = df[['charge_off_aged', 'fee_reversals', 'writeoff_type_aged', 'writeoff_type_null', 'bankruptcy_writeoff_amt']]
```

```python
high_corr
```

```python
high_corr.describe()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

y = df['charge_off']
X_train, X_test, y_train, y_test = train_test_split(high_corr, y, test_size= 0.33)
clf = LogisticRegression().fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = clf.predict(X_test)
accuracy_score(y_pred, y_test)
```

```python
confusion_matrix(y_pred, y_test)
```

```python
# Take out the charge off accounts
charge_off_df = df[df['charge_off'] == 1]
charge_off_df[['snapshot', 'mth_code', 'mob']]
```

```python
count_gb = charge_off_df.groupby("nbr_mths_due").agg("count").reset_index()
count_gb
```

```python
sum_gb = charge_off_df.groupby("mob").agg("sum").reset_index()
sum_gb
```

```python
import matplotlib.pyplot as plt


plt.plot(count_gb['nbr_mths_due'], count_gb['charge_off'])
plt.xlabel("Months due")
plt.ylabel("Number of charge offs")
plt.show()
```

```python
count_gb['recovery_amt']
```

```python
count_gb = charge_off_df.groupby("mob").agg("count").reset_index()
plt.plot(count_gb['mob'], count_gb['charge_off'], color = "red")

plt.xlabel("Months on books")
plt.ylabel("Number of charge offs")
plt.show()

print(count_gb['mob'].min())
```

```python
plt.plot(sum_gb['mob'], sum_gb['recovery_amt'], color = "blue")
plt.xlabel("Months on books")
plt.ylabel("Amount recovered from charge-off")
plt.show()
```
