```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("macro_data.csv")
```

```python
df_date = df[:][7:]
df_date
```

```python
df_date['year'] = df_date['Mnemonic'].str[-4:]
df_date['year'] = pd.to_numeric(df_date['year'])
df_date['M_FGDP.IUSA'] = pd.to_numeric(df_date['M_FGDP.IUSA'])
df_date
```

```python
gdp_gb = df_date.groupby("year").agg("mean").reset_index()
gdp_gb
```

```python
plt.plot(gdp_gb['year'], gdp_gb['M_FGDP.IUSA'])
```
